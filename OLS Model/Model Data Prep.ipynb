```python
df = pd.read_csv("MatchData.csv")
df.info()
```

```python
##Here are some examples of metrics to test
df['GoalsScored']=df['HomeGoal']+df['AwayGoal']
df['HomePassAccuracy']=df['HomePassesCompleted']/df['HomePassesAttempts']
df['HomeShotAccuracy']=df['HomeShotsonTarget']/df['HomeShots']
df['AwayPassAccuracy']=df['AwayPassesCompleted']/df['AwayPassesAttempts']
df['AwayShotAccuracy']=df['AwayShotsonTarget']/df['AwayShots']
```

```python
df.info()
```

```python
df= df.drop(['Match_ID','Wk','Day','Date','Time','Home', 'Score', 'Away','Referee','Match Report','Notes','Venue Key','VAR Key','UtilizationBuckets','HomePossession','AwayPossession'],axis=1)
```

```python
df
```

```python
df['Venue'].value_counts()
```

```python
df=df.groupby('Venue').filter(lambda x : len(x)>29)
```

```python
df['Venue'].value_counts()
```

```python
newdf=df.groupby(['League','Season','Venue']).mean()
```

```python
newdf.reset_index().info()
```

```python
winrate=df.groupby(['League','Season','Venue']).HomeResult.apply(lambda x: (x=='Win').mean())
```

```python
winrate.reset_index().info()
```

```python
joindf=pd.merge(newdf,winrate, how='left', left_on=['League','Season','Venue'], right_on=['League','Season','Venue'])
```

```python
joindf=joindf.reset_index()
```
